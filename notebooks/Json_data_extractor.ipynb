```python
with open('json_data.txt', 'r') as file:
    lines = file.readlines()

# Step 2: Initialize an empty list to store JSON data
json_data_list = []

# Step 3: Iterate through each line in the file
for line in lines:
    # Step 4: Check if 'json_data' is in the line
    if 'json_data' in line:
        # Step 5: Find the index of the earliest '{' after 'json_data'
        start_index = line.find('{')
        # Step 6: If '{' is found, add the substring from '{' to the end of the line to the list
        if start_index != -1:
            json_data_list.append(line[start_index:])

def extract_content(lines):
    extracted_contents = []
    for line in lines:
        content_start = line.find("'content':") + len("'content':")
        role_start = line.find("'role':", content_start)
        content = line[content_start:role_start - 3].strip()
        extracted_contents.append(content)
    return extracted_contents
json_data_list = extract_content(json_data_list)
print(json_data_list)
```

```python
# Filter out items containing "'url': '/embeddings'" from json_data_list
json_data_list = [item for item in json_data_list if "'url': '/embeddings'" not in item]

# Initialize an empty dictionary
item_counts = {}

# Loop through each item in the filtered list
for item in json_data_list:
    # Check if the item is already in the dictionary
    if item in item_counts:
        # Increment the count of the item
        item_counts[item] += 1
    else:
        # Add the item to the dictionary with a count of 1
        item_counts[item] = 1

# Define a function to truncate or summarize the keys for readability
def summarize_key(key, max_length=150):
    # If the key is longer than max_length, truncate and add "..." at the end
    if len(key) > max_length:
        return key[:max_length] + "..."
    else:
        return key

# Print each item and its count in a readable format
for item, count in item_counts.items():
    print(f"{summarize_key(item)}: {count}")

total_count = sum(item_counts.values())
print(f"Total count: {total_count}")
# Calculate the sum of values for keys containing the specified phrase
phrase = "Ensures data integrity, security, and accessibility"
sum_of_values = sum(value for key, value in item_counts.items() if phrase in key)

print(f"Sum of values for keys containing the specified phrase: {sum_of_values}")

```

```python
# Assuming json_data_list is already defined and populated
# Step 1: Convert the list to a string
json_data_str = ''.join(json_data_list)

# Step 2: Calculate the length of the string
length = len(json_data_str)

# Step 3: Find the midpoint of the string
midpoint = length // 10

# Step 4: Print the substring from the beginning up to the midpoint
print(json_data_str[:midpoint])
```

```python
def remove_redundancies(input_file, output_file=None):
    """
    Removes consecutive redundant lines from a text file while preserving the order.

    :param input_file: Path to the input text file.
    :param output_file: Path to the output text file. If None, overwrites the input file.
    """
    if output_file is None:
        output_file = input_file

    with open(input_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Use a loop to remove consecutive duplicate lines
    unique_lines = []
    previous_line = None
    for line in lines:
        if line != previous_line:
            unique_lines.append(line)
            previous_line = line

    with open(output_file, 'w', encoding='utf-8') as file:
        file.writelines(unique_lines)

# Usage
input_path = 'output.txt'
remove_redundancies(input_path)
```
